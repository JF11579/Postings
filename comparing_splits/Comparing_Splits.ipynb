# Now we are only looking 6 styles now: Cape, Colonial,

```python
# Assuming 'data' is your DataFrame containing 'Style', 'Year', and 'Sold_Price' columns

# Group by 'Style' and 'Year', and calculate the median 'Sold_Price' for each group
median_prices = filtered_data.groupby(['Style', 'Year'])['Sold_Price'].median().reset_index()

# Now, 'median_prices' DataFrame contains the median 'Sold_Price' for each style and each year
```

```python
# unique_styles = df['Column_Name'].unique()
median_prices.head(3)
```

```python
# Create a line chart using Plotly Express
fig = px.line(median_prices, x='Year', y='Sold_Price', color='Style',
              title='Median Sold Price by Year and Style')

fig.show()
```

```python
# Sort the DataFrame by 'Style' and 'Year' for consistent calculations
median_prices = median_prices.sort_values(['Style', 'Year'])

# Calculate the percentage difference from one year to the next for each style
median_prices['PercentageChange'] = median_prices.groupby('Style')['Sold_Price'].pct_change()

# The 'PercentageChange' column now contains the percentage change from the previous year to the current year
```

```python
median_prices.head()
```

```python
median_prices.tail()
```

```python
fig = px.line(median_prices, x='Year', y='PercentageChange', color='Style',
              title='Median Sold Price by Year and Style')

fig.show()
```

```python
import plotly.express as px
import plotly.graph_objects as go

# Assuming you have the 'median_prices' DataFrame with 'Year', 'PercentageChange', and 'Style' columns

# Create a line chart using Plotly Express
fig = px.line(median_prices, x='Year', y='PercentageChange', color='Style', title='Median Sold Price Change by Year and Style')

# Add a black horizontal line at y=0
fig.add_shape(go.layout.Shape(
    type="line",
    x0=median_prices['Year'].min(),  # Set the x-coordinate where the line starts
    x1=median_prices['Year'].max(),  # Set the x-coordinate where the line ends
    y0=0,  # Set the y-coordinate of the line
    y1=0,  # Set the y-coordinate of the line
    line=dict(color="black", width=4)  # Set line color and width
))

fig.show()
```

```python
import plotly.express as px
import plotly.graph_objects as go

# Create the violin plot
fig = px.violin(filtered_data, x='Style', y='Sold_Price', color='Style',
                 title='Violin Plot of Sold Price by Style')

# Calculate the maximum kernel density value for each style
max_kde = filtered_data.groupby('Style')['Sold_Price'].apply(lambda x: x.value_counts().max()).reset_index()
max_kde.columns = ['Style', 'Max_KDE']

# Find the Style with the maximum kernel density
max_style = max_kde.loc[max_kde['Max_KDE'].idxmax()]['Style']

# Find the x-coordinate of the maximum kernel density for the chosen style
x_max_kde = filtered_data[filtered_data['Style'] == max_style]['Sold_Price'].value_counts().idxmax()

# Find the y-coordinate of the maximum kernel density
y_max_kde = filtered_data[filtered_data['Style'] == max_style]['Sold_Price'].value_counts().max()

# Add an arrow and annotation
fig.add_shape(
    go.layout.Shape(
        type="line",
        x0=x_max_kde, x1=x_max_kde, xref='x',
        y0=0, y1=y_max_kde, yref='y',
        line=dict(color="black", width=2),
    )
)

fig.add_annotation(
    text=f'Max KDE: {y_max_kde}',
    x=x_max_kde, y=y_max_kde,
    showarrow=True,
    arrowhead=3,
    arrowsize=1,
    arrowwidth=2,
    arrowcolor="black",
    font=dict(size=12),
)

fig.show()

```
