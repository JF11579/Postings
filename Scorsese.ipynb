```python
# Scorsese - movies.csv
data = pd.read_csv('Scorsese - movies.csv')
```

```python
data.head()
```

```python
data = data.sort_values(by= 'Year', ascending=True)
```

```python
data.head()
```

```python
data["x_axis_labels"] = data["Year"].astype(str)

```

# Plot

```python
import plotly.graph_objects as go

# Extract data from the DataFrame
Film = data['Film']
bar_values = data['Minutes']
line_values = data['Tomatoeter']

# Create a bar chart trace
bar_trace = go.Bar(x=Film, y=bar_values, name='Minutes', marker=dict(color='darkgoldenrod'))

# Create a line chart trace
line_trace = go.Scatter(x=Film, y=line_values, mode='lines', name='Rotten Tomatoe Score')

# Create the layout with a black background
layout = go.Layout(
    title='Martin Scorsese Movies: The Length and Score',
    xaxis=dict(title='Year Order'),
    yaxis=dict(title='Minutes'),
    plot_bgcolor='cornflowerblue'
)

# Create the figure and add both traces
fig = go.Figure(data=[bar_trace, line_trace], layout=layout)

# Show the combo plot
fig.show()
```
